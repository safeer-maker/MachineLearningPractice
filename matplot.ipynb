```python
rand = np.random
y = rand.randint(50, size = 50)
x = np.arange(50)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(title="plote of Random Variables", xlabel="Numbers",ylabel= "Random Variables");
```

```python
x = np.linspace(0,10,100)
y = np.random.randint(0,100,100)
y
```

```python
fig,ax = plt.subplots()

ax.scatter (x,np.sin(-x));
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

ax [0,0].plot(x,y)
ax [1,0].scatter(x,np.sin(x))
ax [1,1].hist(x)
ax [0,1].bar(x,np.sin(x))

```
